```python
# This code moves inactive members from the currentMem file to the exMem file and keeps active members in the currentMem file.
# The code updates both files and prints their final contents.

# Dosyaların Açılması:

#currentMem dosyasını r+ modunda açarak hem okuma hem de yazma işlemlerini gerçekleştiriyoruz.
#exMem dosyasını a+ modunda açarak dosyaya ekleme yapabiliyor ve aynı zamanda dosyanın içeriğini okuyabiliyoruz.

#Başlık ve Satırların Okunması:

#currentMem dosyasının başlık satırını okuyoruz ve ardından tüm satırları listeye alıyoruz.
#exMem dosyasının mevcut içeriğini okuyarak oradaki satırları kontrol ediyoruz.

#İnaktif Üyelerin Ayrılması:

#Her bir satırı kontrol ederek 'no' değeri içeren satırları exMem dosyasına ekliyoruz.
#'yes' değeri içeren satırları currentMem dosyasında tutuyoruz.

#Aktif Satırların currentMem Dosyasına Yazılması:

#currentMem dosyasının başına giderek yalnızca aktif üyeleri içeren satırları yazıyoruz.
#Eski içerik varsa, yeni içerik kısa olduğunda dosyanın geri kalan kısmını kesiyoruz.

# Bu güncellenmiş kod ve açıklamalar, dosyalar arasındaki aktif ve inaktif üyeleri doğru bir şekilde ayırmanızı sağlar. 

def cleanFiles(currentMem, exMem):
    
    # currentMem dosyasını okuma ve yazma modunda açıyoruz
    with open(currentMem, 'r+') as current_file:
        # Başlık satırını okuyoruz
        header = current_file.readline()
        
        # Dosyadaki tüm satırları okuyoruz
        lines = current_file.readlines()
        
        
        # exMem dosyasını ekleme ve okuma modunda açıyoruz
        with open(exMem, 'a+') as ex_file:
            # exMem dosyasının başına gidiyoruz
            ex_file.seek(0)
            # exMem dosyasındaki tüm satırları okuyoruz
            ex_lines = ex_file.readlines()
            
            
            # Eğer exMem dosyası boşsa başlığı ekliyoruz
            if not ex_lines:
                ex_file.write(header)
            
            # Her bir satırı dolaşıp inaktif olanları kontrol ediyoruz
            active_lines = [header]  # Aktif üyeler için başlık satırını başlıyoruz
            for line in lines:
                if 'no' in line.split()[-1]:  # 'no' son sütunda mı kontrol ediyoruz
                    ex_file.write(line)  # Inaktif üyeleri exMem dosyasına ekliyoruz
                else:
                    active_lines.append(line)  # Aktif üyeleri saklıyoruz

            current_file.seek(0)
            current_file.writelines(active_lines)
            # Yeni içerik eski içeriğin kısa olup olmadığını kontrol edip dosyanın geri kalan kısmını kesiyoruz
            current_file.truncate()

# Fonksiyonu kullanma örneği (genFiles fonksiyonunun çalıştırılmış olması gerekir)
memReg = '/members.txt'
exReg = '/inactive.txt'
cleanFiles(memReg, exReg)

# Dosyaların son içeriklerini yazdırıyoruz
print("\nAktif Üyeler: \n")
with open(memReg, 'r') as readFile:
    print(readFile.read())

print("\nInaktif Üyeler: \n")
with open(exReg, 'r') as readFile:
    print(readFile.read())
```
